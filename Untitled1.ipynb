```python
from sklearn.datasets import fetch_openml
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the Australian rainfall dataset
data = fetch_openml(name='australian', version=1)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2, random_state=42)

#Scale the data using a StandardScaler
scaler = StandardScaler(with_mean=False)
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Fit a logistic regression model
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred_logreg = logreg.predict_proba(X_test)[:, 1]
auc_logreg = roc_auc_score(y_test, y_pred_logreg)

# Fit a random forest model
rf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict_proba(X_test)[:, 1]
auc_rf = roc_auc_score(y_test, y_pred_rf)

# Fit a gradient boosting model
gb = GradientBoostingClassifier(n_estimators=100, max_depth=3, learning_rate=0.1, random_state=42)
gb.fit(X_train, y_train)
y_pred_gb = gb.predict_proba(X_test)[:, 1]
auc_gb = roc_auc_score(y_test, y_pred_gb)

# Print the AUC scores for each model
print("Logistic Regression AUC:", auc_logreg)
print("Random Forest AUC:", auc_rf)
print("Gradient Boosting AUC:", auc_gb)

# Fit a KNN model
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict_proba(X_test)[:, 1]
auc_knn = roc_auc_score(y_test, y_pred_knn)

# Fit a Decision Tree model
dt = DecisionTreeClassifier(max_depth=5, random_state=42)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict_proba(X_test)[:, 1]
auc_dt = roc_auc_score(y_test, y_pred_dt)

# Fit an AdaBoost model with Decision Tree as base estimator
ada = AdaBoostClassifier(base_estimator=dt, n_estimators=100, learning_rate=0.25, random_state=42)
ada.fit(X_train, y_train)
y_pred_ada = ada.predict_proba(X_test)[:, 1]
auc_ada = roc_auc_score(y_test, y_pred_ada)

# Print the AUC scores for each model
print("KNN AUC:", auc_knn)
print("Decision Tree AUC:", auc_dt)
print("AdaBoost AUC:", auc_ada)
```
